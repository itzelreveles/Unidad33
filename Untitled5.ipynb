Plotting sin linea.
Para trazar soloslos marcadores, puedes utlizar notacion de la cadena de acceso diercto 'o', que signifca "anillos".
Ejemplo: Dibuja dos puntos en el diagrama, uno en la posicion (1,3) y otro en la posicion de (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3,10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

Diagrama de dispercion (scatter)

Se emplean para mostrar la relacion entre los dos variables numericas, detectando correlaciones o agrupamientos
Ejemplo:

```python
import matplotlib.pyplot as plt

altura = [150,160,170,180,190,200]
peso = [50,60,65,80,85,95]

plt.scatter(altura, peso, color =  "red", marker = "x")
plt.title("Relacion entre la altura y el peso")
plt.xlabel("Altura (cms)")
plt.ylabel("Peso (kg)")
plt.show()
```

Graficos de barras (bar y barh)
SIrven para compara categriorias y variables tanto en orientacion vertical como horizontal.
Ejemplo de barras verticales:


```python
import matplotlib.pyplot as plt

categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

plt.bar(categorias, valores, color = "orange")
plt.title("grafica de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
