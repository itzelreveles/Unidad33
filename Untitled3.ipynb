CONCEPTOS BASICOS DE LA MATPLOTLIB

Matplotlib es una libreria de trasados de graficos de bajo nivel en python que sirve como una herramientas de visualizacion, Matplotlib es de codigo abierto y podemos utilizarlos libremente Matplotlib esta escrito principalmente en Python con algunos segmentos en C.
-------------------------------------

INSTALACION DE MATPLOTLIB
-------------------------------
Si tienes en Python y pip ya instalado en tu sistema, instala Matplotlib usando el siguiente comando:
PIP install matplotlib
desde la linea de comando.
----------------------------------------------
IMPORTAR MATPLOTLIB
-------------------------------------------
Una vez instalados matplotlib, importalo en tus aplicasiones agregando el modulo import.
-------------------------------------------------------------------------
ejemplo:

import matplotlib
matplotlib pyplot
-------------------------------------------------------
